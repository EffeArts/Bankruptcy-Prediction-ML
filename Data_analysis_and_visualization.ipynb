### Pre-Processing

```python
# Feature scaling through standardization (or Z-score normalization) can be an important 
# preprocessing step for many machine learning algorithms. Standardization involves rescaling the features  
# such that they have the properties of a standard normal distribution with a mean of zero 
# and a standard deviation of one.
# This makes sure that no one feature greatly dominate the learning model.
# For that we can use StandardScaler.
```

```python
def preProcessing(df):
    df = df.copy()
    
    # Drop single-value column
    df = df.drop(' Net Income Flag', axis=1)
    
    # Split df into X and y
    y = df['Bankrupt?']
    X = df.drop('Bankrupt?', axis=1)
    
    # Train-test split train size 70% and 30% test
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)
    
    # Scale X
    scaler = StandardScaler()
    scaler.fit(X_train)
    X_train = pd.DataFrame(scaler.transform(X_train), index = X_train.index, columns = X_train.columns)
    X_test = pd.DataFrame(scaler.transform(X_test), index = X_test.index, columns = X_test.columns)
    
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = preProcessing(data)
```

# Training (Without Dimentionality reduction)

```python
# Selected models are classification models
original_models = {
    "                   K-Nearest Neighbors": KNeighborsClassifier(),
    "                         Decision Tree": DecisionTreeClassifier(),
    "Support Vector Machine (Linear Kernel)": LinearSVC(),
    "   Support Vector Machine (RBF Kernel)": SVC(),
    "                        Neural Network": MLPClassifier()
}
```

```python
# Trainning the aforementioned models
for name, model in original_models.items():
    model.fit(X_train, y_train)
    print(name + " trained. ")
```

```python
# Get results from every single trainned model.
original_results = []

for name, model in original_models.items():
    result = model.score(X_test, y_test)
    original_results.append(result)
    # showing results as percentages
    print(name + ": {:.2f}%".format(result * 100))
```

### Dimentionality Reduction

#### Dimentionality using PCA

```python
X_train.describe()
```

```python
n_components = 94
pca = PCA(n_components)
pca.fit(X_train)

X_train_reduced = pd.DataFrame(pca.transform(X_train), index = X_train.index, columns=["PC" + str(i) for i in range(1, n_components +1)])
X_test_reduced = pd.DataFrame(pca.transform(X_test), index = X_test.index, columns=["PC" + str(i) for i in range(1, n_components +1)])

# Since it returns an numpy array, the above process puts it back to a DataFrame.
```

```python
X_train_reduced
```

```python
X_train_reduced.describe()
```

```python
# from the above dataset, we can observe that features are now ordered where the ones with the highest standard
# deviation are put on infront/ most variance.
```

```python
# Explained variance

pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
# Now let's plot this, for better Vizualization.

fig = px.bar(
    x=["PC" + str(i) for i in range(1, n_components + 1)],
    y=pca.explained_variance_ratio_,
    labels={'x': "Principal Component", 'y': "Variance Ratio"},
    color=pca.explained_variance_ratio_,
    color_continuous_scale=[(0, 'lightblue'), (1, 'darkblue')],
    title="Proportion of Variance in Principal Components"
)

fig.show()
```

```python
# It can be seen that the features on the far right side of the diagram, contains very little information,
# very little variance.
# With dimnetionality reduction, they will get dropped, because they are not as useful to us as the higher columns.
# We might reduce it to the first ten features. Yes, we might loose some information, but atleast we will
# have most of it, since it is captured in the first features which have a high variation.
# this is a very usefull method to use, especially when dealing with data with high dimentionality.
# Like in our case.
```

```python
# Let's drop it to like 10 of them. Just keep 10 of them.
```

```python
n_components = 25
pca = PCA(n_components)
pca.fit(X_train)

X_train_reduced = pd.DataFrame(pca.transform(X_train), index = X_train.index, columns=["PC" + str(i) for i in range(1, n_components +1)])
X_test_reduced = pd.DataFrame(pca.transform(X_test), index = X_test.index, columns=["PC" + str(i) for i in range(1, n_components +1)])

# Since it returns an numpy array, the above process puts it back to a DataFrame.

# Now let's plot this, for better Vizualization.

fig = px.bar(
    x=["PC" + str(i) for i in range(1, n_components + 1)],
    y=pca.explained_variance_ratio_,
    labels={'x': "Principal Component", 'y': "Variance Ratio"},
    color=pca.explained_variance_ratio_,
    color_continuous_scale=[(0, 'lightblue'), (1, 'darkblue')],
    title="Proportion of Variance in Principal Components"
)

fig.show()
```

## Training (Reduced Data)

```python
# Selected models are classification models
reduced_models = {
    "                   K-Nearest Neighbors": KNeighborsClassifier(),
    "                         Decision Tree": DecisionTreeClassifier(),
    "Support Vector Machine (Linear Kernel)": LinearSVC(),
    "   Support Vector Machine (RBF Kernel)": SVC(),
    "                        Neural Network": MLPClassifier()
}

# Trainning the aforementioned models
for name, model in reduced_models.items():
    model.fit(X_train_reduced, y_train)
    print(name + " trained. ")
```

```python
reduced_results = []

for name, model in reduced_models.items():
    result = model.score(X_test_reduced, y_test)
    reduced_results.append(result)
    # showing results as percentages
    print(name + ": {:.2f}%".format(result * 100))
```

### Change in performance after Dimentionality Reduction with PCA

```python
fig = px.bar(
    x=np.subtract(reduced_results, original_results),
    y=original_models.keys(),
    orientation='h',
    labels={'x': "Change in Performance", 'y': "Model"},
    color=np.subtract(reduced_results, original_results),
    color_continuous_scale=[(0, 'red'), (1, 'blue')],
    title="Change in Model Performance After Dimensionality Reduction"
)

fig.show()
```
